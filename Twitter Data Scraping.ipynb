# Extract all the tweets from Twitter

```python
def extractTweets(setSince,setUntil,text_query):
    language = 'en'
    
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query).setSince(setSince).setUntil(setUntil).setEmoji("unicode").setLang(language)
    # Creation of list that contains all tweets
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    # Creating list of chosen tweet data
    text_tweets = [[tweet.id, tweet.permalink,  tweet.to, tweet.text, tweet.date, tweet.retweets,
                        tweet.mentions, tweet.hashtags] for tweet in tweets]
    
    tweets_df = pd.DataFrame(text_tweets, columns = ['TweetID', 'Permalink','To','Text','Date','Retweets',
                                                 'Mentions','Hashtags'])
        
    return tweets_df

All_tweets_df=pd.DataFrame(columns = ['TweetID', 'Permalink','To','Text','Date','Retweets','Mentions','Hashtags'])

def DateCreation(i):
    today = date.today()
    DD = timedelta(days=i)
    earlier = today - DD
    Earlier_Date = earlier.strftime("%Y-%m-%d")
    return Earlier_Date

NoP=4
NoD=60
start=time.time()
for i in range(NoP):
    tweets_df1 = extractTweets(DateCreation((i+1)*NoD-1), DateCreation(i*NoD),'dimentia COVID -Donald -Trump -Joe -Biden -president -Tom -Seaver -donaldtrump -joebiden -tomseaver')
    tweets_df2 = extractTweets(DateCreation((i+1)*NoD-1), DateCreation(i*NoD),'dimentia corona -Donald -Trump -Joe -Biden -president -Tom -Seaver -donaldtrump -joebiden -tomseaver')
    All_tweets_df = pd.concat([All_tweets_df,tweets_df1,tweets_df2])
    time.sleep(600)
end=time.time()
ElapsedTime=end-start
print(ElapsedTime)
```

# Data Preprocessing

```python
def addAND(data):
    '''Replace '&amp' with 'and' '''
    
    data.drop_duplicates(['Text'], inplace=True)
    data['Text'] = data['Text'].replace({'&amp;': ' and '}, regex=True)
#     data['Text'] = data['Text'].replace({'â€™': "\'"}, regex=True) 
    return data

def removeHTTP(data):
    '''Remove tweets having link (https)'''
    mask = np.column_stack(data[col].str.contains('http', na=False) for col in data[['Text']])
    http = data.loc[mask.any(axis=1)]
    data = data[~data.Text.isin(http.Text)]
    return data

def removeNames(data):
    '''Remove any Trump, Biden, or Seaver related tweets'''
    for word in ['president', 'joebiden', 'donaldtrump', 'trump', 'biden', 'sincerely', 'tom', 'seaver', 'tomseaver']:
        mask = np.column_stack(data[col].str.lower().str.contains(word, na=False) for col in data[['To', 'Text', 
                                                                                            'Mentions', 'Hashtags']])
        shouldRemoved = data.loc[mask.any(axis=1)]
        data = data[~data.Text.isin(shouldRemoved.Text)]
    return data

def nonReplies(stories):
    replies = [row for row in stories['Text'] if row.startswith('@') or row.startswith('.@') or row.startswith('. @')]
    non_replies = stories[~stories.Text.isin(replies)]
    return non_replies

def extractStories(data):
    '''Find Stroies with specific words'''
    Stories = pd.DataFrame(columns = ['TweetID', 'Permalink','To','Text','Date','Retweets','Mentions','Hashtags'])

    for word in ['my', 'mine', 'family', 'father', 'mother', 'sister', 'bother', 'friend', 'dad', 'mom', 'mum', 'parents']:
        mask = np.column_stack(data[col].str.lower().str.contains(word, na=False) for col in data[['Text']])
        shouldRemoved = data.loc[mask.any(axis=1)]
        Stories = Stories.append(shouldRemoved, sort = True)
        
    Stories.drop_duplicates(['Text'],inplace = True) 
    CleanStories = nonReplies(Stories)
    CleanStories.to_csv("Data/Stories.csv", index=False )
    CleanStories['Date'] = CleanStories['Date'].astype(str)
    CleanStories.to_excel("Data/Stories.xlsx", index=False)
    return CleanStories

def extractNews(allData, stories):
    '''Extract non-stories'''
    News = allData[~allData.Text.isin(stories.Text)]
    News.to_csv("Data/News.csv", index=False)
    return News


dataset = addAND(All_tweets_df)
print(f"Lenght of All data: {len(All_tweets_df)}")
dataset = removeHTTP(dataset)
print(f"Lenght of All Data after removing https: {len(dataset)}")
dataset = removeNames(dataset)
print(f"Lenght of Cleaned Dataset: {len(dataset)}")
Stories = extractStories(dataset)
print(f"Lenght of Stories: {len(Stories)}")
News = extractNews(dataset, Stories)
print(f"Lenght of non-stories: {len(News)}")
CleanStories = nonReplies(Stories)
print(f"Lenght of non-replies: {len(CleanStories)}")
```

```python
import docx
import csv

doc = docx.Document()

with open('Data/CleanedStories.csv', newline='', encoding="utf8") as f:
    csv_reader = csv.reader(f) 

    csv_headers = next(csv_reader)
    csv_cols = len(csv_headers)

    table = doc.add_table(rows=2, cols=csv_cols)
    hdr_cells = table.rows[0].cells

    for i in range(csv_cols):
        hdr_cells[i].text = csv_headers[i]

    for row in csv_reader:
        row_cells = table.add_row().cells
        for i in range(csv_cols):
            row_cells[i].text = row[i]

doc.add_page_break()
doc.save("Data/Stories.docx")
```

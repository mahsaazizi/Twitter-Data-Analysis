```python
import GetOldTweets3 as got
import pandas as pd
from datetime import date, datetime, timedelta
import time
import numpy as np
import seaborn as sn
```

# Extract all the tweets from Twitter

```python
def extractTweets(setSince,setUntil,text_query):
    language = 'en'
    
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query).setSince(setSince).setUntil(setUntil).setEmoji("unicode").setLang(language)
    # Creation of list that contains all tweets
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    # Creating list of chosen tweet data
    text_tweets = [[tweet.id, tweet.permalink, tweet.username, tweet.to, tweet.text, tweet.date, tweet.retweets, tweet.favorites
                        ,tweet.mentions, tweet.hashtags, tweet.geo] for tweet in tweets]
    tweets_df = pd.DataFrame(text_tweets, columns = ['TweetID', 'Permalink','Username','To','Text','Date','Retweets',
                                                 'Favorites','Mentions','Hashtags','Geo'])
        
    return tweets_df

All_tweets_df=pd.DataFrame(columns = ['TweetID', 'Permalink','Username','To','Text','Date','Retweets',
                                                 'Favorites','Mentions','Hashtags','Geo'])

def DateCreation(i):
    today = date.today()
    DD = timedelta(days=i)
    earlier = today - DD
    Earlier_Date = earlier.strftime("%Y-%m-%d")
    return Earlier_Date

NoP=4
NoD=60
start=time.time()
for i in range(NoP):
    tweets_df1 = extractTweets(DateCreation((i+1)*NoD-1), DateCreation(i*NoD),'dimentia COVID -Donald -Trump -Joe -Biden -president')
    tweets_df2 = extractTweets(DateCreation((i+1)*NoD-1), DateCreation(i*NoD),'dimentia corona -Donald -Trump -Joe -Biden -president')
    All_tweets_df = pd.concat([All_tweets_df,tweets_df1,tweets_df2])
    time.sleep(600)
end=time.time()
ElapsedTime=end-start
print(ElapsedTime)
```

# Read all the stored data

```python
All_tweets_df = pd.read_csv('Data/All_tweets_df.csv')
print(len(All_tweets_df))
All_tweets_df.drop_duplicates(['Text'],inplace = True)
print(len(All_tweets_df))
```

# Remove any Trump related tweets

```python
CleanedTweets = All_tweets_df.copy()
sum = 0
for word in ['Trump', 'Biden', 'President', 'JoeBiden', 'DonaldTrump', 'trump', 'biden', 'Sincerely']:
    mask = np.column_stack(CleanedTweets[col].str.contains(word, na=False) for col in CleanedTweets[['Username', 'To', 'Text', 'Mentions', 'Hashtags']])
    shouldRemoved = CleanedTweets.loc[mask.any(axis=1)]
    print(len(shouldRemoved))
    sum = sum + len(shouldRemoved)
    CleanedTweets = CleanedTweets[~CleanedTweets.Text.isin(shouldRemoved.Text)]
print('Total filtered Data: ', sum)
```

# Replace '&amp' with ' and ' 

```python
CleanedTweets['Text'] = CleanedTweets['Text'].replace({'&amp;': ' and '}, regex=True)
```

```python
Stories = pd.DataFrame(columns = ['TweetID', 'Permalink','Username','To','Text','Date','Retweets',
                                                 'Favorites','Mentions','Hashtags','Geo'])
sum = 0
# for item in ['Username', 'To', 'Text', 'Mentions', 'Hashtags']:
#     CleanedTweets[item] = CleanedTweets[item].str.lower()
# lowercaseData = CleanedTweets[[]].str.lower()
for word in ['my', 'mine', 'family', 'father', 'mother', 'sister', 'bother', 'friend', 'dad', 'mom', 'mum', 'parents',
            'My', 'Mine', 'Family', 'Father', 'Mother', 'Sister', 'Bother', 'friend', 'Dad', 'Mom', 'Mum', 'Parents']:
    mask = np.column_stack(CleanedTweets[col].str.contains(word, na=False) for col in CleanedTweets[['Text']])
    shouldRemoved = CleanedTweets.loc[mask.any(axis=1)]
    print(len(shouldRemoved))
    sum = sum + len(shouldRemoved)
    Stories = Stories.append(shouldRemoved, sort = True)
print('Total story Data: ', sum)
Stories.drop_duplicates(['Text'],inplace = True)    
print(len(Stories))
```
